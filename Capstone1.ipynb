```python
import matplotlib.pyplot as plt
import seaborn as sns

# Events by State and Year
events_by_state_year = data.groupby(['State', data['Date'].dt.year]).size().unstack(fill_value=0)
print("\nEvents by State and Year:")
print(events_by_state_year.head(10))  # Top 10 states for brevity

# Heatmap for events by state and year (top states)
top_states = data['State'].value_counts().head(10).index
plt.figure(figsize=(12, 8))
sns.heatmap(events_by_state_year.loc[top_states], cmap='Blues', annot=True, fmt='d')
plt.title('Tesla Accident Events by State and Year (Top 10 States)')
plt.xlabel('Year')
plt.ylabel('State')
plt.show()

```

```python
# Events by Country and Year
events_by_country_year = data.groupby(['Country', data['Date'].dt.year]).size().unstack(fill_value=0)
print("\nEvents by Country and Year:")
print(events_by_country_year)

# Heatmap for events by country and year
plt.figure(figsize=(12, 6))
sns.heatmap(events_by_country_year, cmap='Greens', annot=True, fmt='d')
plt.title('Tesla Accident Events by Country and Year')
plt.xlabel('Year')
plt.ylabel('Country')
plt.show()

```

```python
# Events by Month per State (example for top states)
events_by_state_month = data.groupby(['State', data['Date'].dt.month]).size().unstack(fill_value=0)
print("\nEvents by State and Month (Top 10 States):")
print(events_by_state_month.loc[top_states])

# Heatmap for events by state and month
plt.figure(figsize=(12, 8))
sns.heatmap(events_by_state_month.loc[top_states], cmap='Oranges', annot=True, fmt='d')
plt.title('Tesla Accident Events by State and Month (Top 10 States)')
plt.xlabel('Month')
plt.ylabel('State')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()

```

```python
# Events by Day per State/Country: Too granular for plot; use table (e.g., for a specific state)
# Example for USA
usa_data = data[data['Country'] == 'USA']  # Adjust country code if needed
events_by_day_usa = usa_data.groupby(usa_data['Date'].dt.date).size()
print("\nDaily Events in USA (Sample):")
print(events_by_day_usa.head(10))
```

```python
# Number of victims (deaths) in each accident
print("\nSummary of Deaths per Event:")
print(data['Deaths'].describe())

# Plot distribution
plt.figure(figsize=(8, 6))
sns.histplot(data['Deaths'], bins=20, kde=True, color='purple')
plt.title('Distribution of Deaths per Tesla Event')
plt.xlabel('Number of Deaths')
plt.ylabel('Frequency')
plt.show()

```

```python
# How many times did Tesla drivers die? (Total driver deaths across all events)
tesla_driver_deaths = data['Tesladriver'].sum()
print(f"\nTotal Tesla Driver Deaths: {tesla_driver_deaths}")

# Proportion of events with one or more occupants died (driver + occupant)
events_with_occupant_deaths = data[(data['Tesladriver'] > 0) | (data['Teslaoccupant'] > 0)].shape[0]
total_events = data.shape[0]
proportion = events_with_occupant_deaths / total_events
print(f"Proportion of Events with 1+ Tesla Occupant/Driver Deaths: {proportion:.2%}")

# Distribution of events hitting cyclist/pedestrian (number of such deaths per event)
print("\nDistribution of Cyclist/Pedestrian Deaths:")
print(data['Cyclists_Peds'].value_counts())

# Plot
plt.figure(figsize=(8, 6))
sns.histplot(data['Cyclists_Peds'], bins=10, kde=False, color='red')
plt.title('Distribution of Cyclist/Pedestrian Deaths in Events')
plt.xlabel('Number of Cyclist/Pedestrian Deaths')
plt.ylabel('Number of Events')
plt.show()

```

```python
# How many times accident involved death of Tesla occupant/driver along with cyclist/pedestrian? (Number of such events)
tsla_cycl_peds_events = data[data['TSLA_cycl_peds'] > 0].shape[0]
print(f"Events with Tesla Occupant/Driver Death + Cyclist/Pedestrian Death: {tsla_cycl_peds_events}")

# Frequency of Tesla colliding with other vehicles (events with other vehicle deaths)
vehicle_collision_events = data[data['Othervehicle'] > 0].shape[0]
print(f"Events Involving Collision with Other Vehicles (Based on Deaths): {vehicle_collision_events}")

# Plot
plt.figure(figsize=(6, 4))
sns.countplot(x=(data['Othervehicle'] > 0), data=data, palette='Set1')
plt.title('Events with Other Vehicle Deaths')
plt.xlabel('Other Vehicle Involved (Deaths > 0)')
plt.ylabel('Number of Events')
plt.show()
```

```python
# Events by Model
events_by_model = data.groupby('Model').size().sort_values(ascending=False)
print("\nEvents by Model:")
print(events_by_model)

# Plot events by model
plt.figure(figsize=(10, 6))
events_by_model.plot(kind='bar', color='goldenrod')
plt.title('Number of Tesla Death Events by Model')
plt.xlabel('Model')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.show()

# Optional: Pie chart for proportion
plt.figure(figsize=(8, 8))
events_by_model.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Tesla Death Events by Model')
plt.ylabel('')  # Remove y-label for pie chart
plt.show()
```

```python
# Check unique values in VTAD to confirm it's binary or count
print("\nUnique Values in VTAD:")
print(data['VTAD'].unique())

# Number of verified Autopilot death events (where VTAD > 0)
verified_events = data[data['VTAD'] > 0].shape[0]
total_events = data.shape[0]
proportion_verified = verified_events / total_events if total_events > 0 else 0
print(f"\nNumber of Verified Tesla Autopilot Death Events: {verified_events}")
print(f"Proportion of Verified Events: {proportion_verified:.2%}")

# Events by Model for verified Autopilot deaths
verified_by_model = data[data['VTAD'] > 0].groupby('Model').size().sort_values(ascending=False)
print("\nVerified Autopilot Death Events by Model:")
print(verified_by_model)

plt.figure(figsize=(10, 6))
verified_by_model.plot(kind='bar', color='teal')
plt.title('Verified Tesla Autopilot Death Events by Model')
plt.xlabel('Model')
plt.ylabel('Number of Events')
plt.xticks(rotation=45)
plt.show()
```

```python
# Total deaths in verified events
total_deaths_verified = data[data['VTAD'] > 0]['Deaths'].sum()
print(f"Total Deaths in Verified Autopilot Events: {total_deaths_verified}")

# Distribution of deaths in verified events
print("\nSummary of Deaths in Verified Autopilot Events:")
print(data[data['VTAD'] > 0]['Deaths'].describe())

# Plot distribution of deaths in verified events
plt.figure(figsize=(8, 6))
sns.histplot(data[data['VTAD'] > 0]['Deaths'], bins=10, kde=True, color='darkblue')
plt.title('Distribution of Deaths in Verified Tesla Autopilot Events')
plt.xlabel('Number of Deaths')
plt.ylabel('Frequency')
plt.show()
```

```python
# Proportion plot (verified vs non-verified)
plt.figure(figsize=(6, 6))
sns.countplot(x='VTAD', data=data, palette='Set2', hue='VTAD', legend=False)  # Use 'data' and fix palette usage
plt.title('Distribution of Verified vs. Non-Verified Autopilot Death Events')
plt.xlabel('Verified Autopilot (1 = Yes)')
plt.ylabel('Number of Events')
plt.show()
```

```python
# Optional: Verified events by year
verified_by_year = data[data['VTAD'] > 0].groupby(data['Date'].dt.year).size()
print("\nVerified Autopilot Events by Year:")
print(verified_by_year)

plt.figure(figsize=(10, 6))
verified_by_year.plot(kind='bar', color='navy')
plt.title('Verified Tesla Autopilot Death Events by Year')
plt.xlabel('Year')
plt.ylabel('Number of Events')
plt.show()
```
